```python
# Order class
class Order:
    def __init__(self, orderID, refrenceNum, deliveryDate, deliveryMethod, weight): # Attributes that have the order details
        self.__orderID = orderID  # Unique order identifier
        self.__refrenceNum = refrenceNum  # Reference number of the order
        self.__deliveryDate = deliveryDate  # Expected delivery date
        self.__deliveryMethod = deliveryMethod  # Method used for delivery (e.g., Courier)
        self.__weight = weight  # Weight of the order in kg

    # Setter and getter methods for order ID
    def setOrderID(self, orderID):
        self.__orderID = orderID
    def getOrderID(self):
        return self.__orderID # Returns the order ID

    # Setter and getter for reference number
    def setRefrenceNum(self, refrenceNum):
        self.__refrenceNum = refrenceNum
    def getRefrenceNum(self):
        return self.__refrenceNum # Returns the refrence number of the order

    # Setter and getter for delivery date
    def setDeliveryDate(self, deliveryDate):
        self.__deliveryDate = deliveryDate
    def getDeliveryDate(self):
        return self.__deliveryDate # Returns the delivery date of the order

    # Setter and getter for delivery method
    def setDeliveryMethod(self, deliveryMethod):
        self.__deliveryMethod = deliveryMethod
    def getDeliveryMethod(self):
        return self.__deliveryMethod # Returns the delivery method of the order

    # Setter and getter for weight
    def setWeight(self, weight):
        self.__weight = weight
    def getWeight(self):
        return self.__weight # Returns the order's weight

    # Placeholder function that can cancel orders
    def cancelOrder(self):
        pass

# Class for user details
class User:
    def __init__(self, firstName, lastName, email, contactNum, deliveryAd):
        self.__firstName = firstName  # User's first name
        self.__lastName = lastName  # User's last name
        self.__email = email  # Email address
        self.__contactNum = contactNum  # Contact number
        self.__deliveryAd = deliveryAd  # Delivery address

    # Getter and setter methods for first name
    def setFirstName(self, firstName):
        self.__firstName = firstName
    def getFirstName(self):
        return self.__firstName # Returns the first name

    # Getter and setter methods for last name
    def setLastName(self, lastName):
        self.__lastName = lastName
    def getLastName(self):
        return self.__lastName # Returns the last name

    # Email setter and getter
    def setEmail(self, email):
        self.__email = email
    def getEmail(self):
        return self.__email # Returns the user's email

    # Contact Number setter and getter
    def setContactNum(self, contactNum):
        self.__contactNum = contactNum
    def getContactNum(self):
        return self.__contactNum # Returns the user's contact number

    # Setter and getter for delivery address
    def setDeliveryAd(self, deliveryAd):
        self.__deliveryAd = deliveryAd
    def getDeliveryAd(self):
        return self.__deliveryAd # Returns the user's delivery adddress

# Employee class
class Employee:
    def __init__(self, employeeID, firstName, lastName,): # Attributes that have the employee details
        self.__employeeID = employeeID
        self.__firstName = firstName
        self.__lastName = lastName

    # Setter and getter for Employee ID
    def setEmployeeID(self, employeeID):
        self.__employeeID = employeeID
    def getEmployeeID(self):
        return self.__employeeID
  # Getter and setter methods for first name
    def setFirstName(self, firstName):
        self.__firstName = firstName
    def getFirstName(self):
        return self.__firstName # Returns the first name

    # Getter and setter methods for last name
    def setLastName(self, lastName):
        self.__lastName = lastName
    def getLastName(self):
        return self.__lastName # Returns the last name

# Class for the Delivery Worker
class DeliveryWorker:
    def __init__(self, workerID, firstName, lastName,): # Attributes that have the delivery worker details
        self.__workerID = workerID
        self.__firstName = firstName
        self.__lastName = lastName

    # Getter and setter methofs for the Delivery worker's ID
    def setWorkerID(self, workerID):
        self.__workerID = workerID
    def getWorkerID(self):
        return self.__workerID

  # Getter and setter methods for first name
    def setFirstName(self, firstName):
        self.__firstName = firstName
    def getFirstName(self):
        return self.__firstName # Returns the first name

    # Getter and setter methods for last name
    def setLastName(self, lastName):
        self.__lastName = lastName
    def getLastName(self):
        return self.__lastName # Returns the last name

# Delivery Note class
class DeliveryNote:
    def __init__(self, itemCode, quantity, price, taxesFees): # Attributes of the delivery note detials
        self.__itemCode = itemCode
        self.__quantity = quantity
        self.__price = price
        self.__totalPrice = 0  # Will be calculated
        self.__taxesFees = taxesFees

 # Setter and getter for itemCode
    def setItemCode(self, itemCode):
        # Sets the item code.
        self.__itemCode = itemCode

    def getItemCode(self):
        # Returns the item code.
        return self.__itemCode

    # Setter and getter for quantity
    def setQuantity(self, quantity):
        # Sets the item quantity.
        self.__quantity = quantity

    def getQuantity(self):
        # Returns the item quantity.
        return self.__quantity

    # Setter and getter for price
    def setPrice(self, price):
        # Sets the price per unit of the item.
        self.__price = price

    def getPrice(self):
        # Returns the price per unit of the item.
        return self.__price

    # Setter and getter for taxes and fees
    def setTaxesFees(self, taxesFees):
        # Sets the taxes and fees applicable to the order.
        self.__taxesFees = taxesFees

    def getTaxesFees(self):
        # Returns the taxes and fees applicable to the order.
        return self.__taxesFees

    # Calculates total price before taxes
    def calculateTotalPrice(self):
        self.__totalPrice = self.__quantity * self.__price
        return self.__totalPrice # Returns the total price

   # Calculates final total charges including taxes
    def calculateTotalCharges(self):
        if self.__totalPrice == 0:
            self.calculateTotalPrice()  # Ensure total price is calculated first
        return self.__totalPrice + self.__taxesFees # Returns the total price added with the taxes

    # This functions displays the items code, quantity, price of each item, and total price
    def displayItem(self):
      return f"{self.__itemCode} {self.__quantity}  {self.__price:.2f}  {self.calculateTotalPrice():.2f}"


# Creating object for User
user1 = User("Sarah", "Johnson", "sarah.johnson@example.com", 97155886789, "45 Knowledge Avenue, Dubai, UAE")

# Creating object for Employee
employee1 = Employee(101, "Williams", "Smith")

# Creating object for Delivery Worker
worker1 = DeliveryWorker(201, "David", "Brown")

# Creating object for Order
order1 = Order("DEL123456789", "DN-2025-001", "February 28, 2025", "Courier", 7.0)

# Creating Delivery Note Items
item1 = DeliveryNote("ITM001", 1, 100.00, 5.00)
item2 = DeliveryNote("ITM002", 1, 75.00, 3.75)
item3 = DeliveryNote("ITM003", 1, 120.00, 6.00)
item4 = DeliveryNote("ITM004", 3, 15.00, 2.75)

# Calculating total prices
total1 = item1.calculateTotalPrice()
total2 = item2.calculateTotalPrice()
total3 = item3.calculateTotalPrice()
total4 = item4.calculateTotalPrice()

# Calculating total amounts
subtotal = total1 + total2 + total3 + total4
totalTaxesFees = item1.getTaxesFees() + item2.getTaxesFees() + item3.getTaxesFees() + item4.getTaxesFees()
totalCharges = subtotal + totalTaxesFees

# Displaying the Delivery Note
print("=" * 60)
print(" " * 20 + "DELIVERY NOTE")
print("=" * 60)

# Recipient Details
print("\nRecipient Details:")
print(f"Name: {user1.getFirstName()} {user1.getLastName()}")
print(f"Email: {user1.getEmail()}")
print(f"Contact Number: +{user1.getContactNum()}")
print(f"Delivery Address: {user1.getDeliveryAd()}")

# Delivery Information
print("\nDelivery Information:")
print(f"Order Number: {order1.getOrderID()}")
print(f"Reference Number: {order1.getRefrenceNum()}")
print(f"Delivery Date: {order1.getDeliveryDate()}")
print(f"Delivery Method: {order1.getDeliveryMethod()}")
print(f"Handled by Employee: {employee1.getFirstName()} {employee1.getLastName()}")
print(f"Delivery Worker: {worker1.getFirstName()} {worker1.getLastName()}")
print(f"Total Weight: {order1.getWeight()} kg")

# Summary of Items Delivered
print("\nSummary of Items Delivered:")
print(f"{'Item Code':<10} {'Description':<30} {'Quantity':<10} {'Unit Price (AED)':<20} {'Total Price (AED)':<20}")
print("-" * 80)
print(f"{item1.getItemCode():<10} {'Wireless Keyboard':<30} {item1.getQuantity():<10} {item1.getPrice():<20.2f} {total1:<20.2f}")
print(f"{item2.getItemCode():<10} {'Wireless Mouse & Pad Set':<30} {item2.getQuantity():<10} {item2.getPrice():<20.2f} {total2:<20.2f}")
print(f"{item3.getItemCode():<10} {'Laptop Cooling Pad':<30} {item3.getQuantity():<10} {item3.getPrice():<20.2f} {total3:<20.2f}")
print(f"{item4.getItemCode():<10} {'Camera Lock':<30} {item4.getQuantity():<10} {item4.getPrice():<20.2f} {total4:<20.2f}")
print("-" * 80)

# Total Summary
print(f"\nSubtotal: AED {subtotal:.2f}")
print(f"Taxes and Fees: AED {totalTaxesFees:.2f}")
print(f"Total Charges: AED {totalCharges:.2f}")

```
